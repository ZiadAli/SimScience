```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Person:

  def __init__(self, x, y):
    self.x = x
    self.y = y
```

```python
# Instantiate a person (call them human1) with set coordinates so we can plot them

human1 = Person(5, 3)
```

```python
# Plot the person in the X-Y plane using the function below

plt.scatter(human1.x, human1.y)
```

```python
# Create 100 people (without copy-pasting your code) and plot each of them on the same graph 

for i in range(100):
  human = Person(np.random.random(), np.random.random())
  plt.scatter(human.x, human.y)
```

```python
#   Create all 100 people before plotting them (hint: store their location coordinates in a data structure)
humans = []
humans_x = []
humans_y = []

for i in range(100):
  x = np.random.random()
  y = np.random.random()
  human = Person(x, y)

  humans.append(human)
  humans_x.append(x)
  humans_y.append(y)

plt.scatter(humans_x, humans_y)
```

```python
#   Add a parameter to the Person class so you can control the color of each individual person when they're plotted (hint: Use plt.scatter's 'color' parameter)

class Person2:

  def __init__(self, x, y, color):
    self.x = x
    self.y = y
    self.color = color

humans = []
humans_x = []
humans_y = []
humans_colors = []

color = "green"
for i in range(100):
  x = np.random.random()
  y = np.random.random()
  human = Person2(x, y, color)

  humans.append(human)
  humans_x.append(x)
  humans_y.append(y)
  humans_colors.append(color)

plt.scatter(humans_x, humans_y, color = humans_colors)
```

```python
class Car:
  def __init__(self, make, model, year, gas):
    self.make = make
    self.model = model
    self.year = year
    self.gas = gas
    
```

```python
ziad_car = Car('Toyota', 'Camry', 1999, 40)
```

```python
print("Ziad's Car's Gas: ", ziad_car.gas)
ziad_car.gas = ziad_car.gas - 5
print("Ziad's Car's Gas: ", ziad_car.gas)
```

```python
miles_driven = 20
ziad_car.gas = ziad_car.gas - 1 - miles_driven*0.3
print("Ziad's Car's Gas: ", ziad_car.gas)
```

```python
class Car:
  def __init__(self, make, model, year, gas):
    self.make = make
    self.model = model
    self.year = year
    self.gas = gas

  def drive(self, miles):
    self.gas = self.gas - 1 - miles*0.3
```

```python
ziad_car.drive(miles_driven)
print("Ziad's Car's Gas: ", ziad_car.gas)
```
